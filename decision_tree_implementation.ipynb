```python
# Load the necessary libraries
import yfinance as yf
import talib as ta
import numpy as np

# Fetch BTC price data
data = yf.download('BTC-USD')
data
```

```python
data['EMA10'] = ta.EMA(data['Adj Close'].values, timeperiod=10)
data['EMA30'] = ta.EMA(data['Adj Close'].values, timeperiod=30)
data['ATR'] = ta.ATR(data['High'].values, data['Low'].values,
                    data['Adj Close'].values, timeperiod=14)
data['ADX'] = ta.ADX(data['High'].values, data['Low'].values,
                    data['Adj Close'].values, timeperiod=14)
data['RSI'] = ta.RSI(data['Adj Close'].values, timeperiod=14)

macd, macdsignal, macdhist = ta.MACD(data['Adj Close'].values, 
                                    fastperiod=12, slowperiod=26, 
                                     signalperiod=9)
data['MACD'] = macd
data['MACDsignal'] = macdsignal
data.tail()
```

```python
import matplotlib.pyplot as plt

# Plot BTC Price
plt.figure(figsize=(10, 7))
plt.title('BTC Close Price')
plt.ylabel('Price')
plt.plot(data['Adj Close'][-100:-1], label='Adj Close')
plt.plot(data['EMA10'][-100:-1], label='EMA10')
plt.plot(data['EMA30'][-100:-1], label='EMA30')
plt.legend()
```

```python
data['ClgtEMA10'] = np.where(data['Adj Close'] > data['EMA10'], 1, -1)
data['EMA10gtEMA30'] = np.where(data['EMA10'] > data['EMA30'], 1, -1)
data['MACDSIGgtMACD'] = np.where(data['MACDsignal'] > data['MACD'], 1, -1)
```

```python
# Creating the target variables
data['Return'] = data['Close'].pct_change(1).shift(-1)
data['target_cls'] = np.where(data.Return > 0, 1, 0)
data['target_rgs'] = data['Return']
data.tail()
```

```python
# Obtaining the data set for decision trees
predictors_list = ['ATR', 'ADX', 'RSI', 'ClgtEMA10', 'EMA10gtEMA30', 'MACDSIGgtMACD']
X = data[predictors_list]
X.tail()
```

```python
X.isnull().sum()
```

```python
X = X.iloc[27:-1, :]
```

```python
X.describe()
```

```python
y_cls = data.target_cls.iloc[27: -1]
y_cls
```

```python
y_rgs = data.target_rgs.iloc[27:-1]
y_rgs
```

```python
# Splitting the data into training and testing data sets
from sklearn.model_selection import train_test_split

y = y_cls
X_cls_train, X_cls_test, y_cls_train, y_cls_test = train_test_split(X, y, test_size=0.3, random_state=432, stratify=y)

print(X_cls_train.shape, y_cls_train.shape)
print(X_cls_test.shape, y_cls_test.shape)
```

```python
train_length = int(len(X)*0.7)
X_rgs_train = X[:train_length]
X_rgs_test = X[train_length:]
y_rgs_train = y_rgs[:train_length]
y_rgs_test = y_rgs[train_length:]

print(X_rgs_train.shape, y_rgs_train.shape)
print(X_rgs_test.shape, y_rgs_test.shape)
```

```python
# Decision Trees for Classification
from sklearn.tree import DecisionTreeClassifier as DTC

clf = DTC(criterion='gini', max_depth=3, min_samples_leaf=6)
clf
```

```python
clf = clf.fit(X_cls_train, y_cls_train)
clf
```

```python
# Visualize Decision Trees for Classification
from sklearn import tree
import graphviz

dot_data = tree.export_graphviz(clf, out_file=None, filled=True,
                               feature_names=predictors_list)
graphviz.Source(dot_data)
```

```python
# Make forecast
y_cls_pred = clf.predict(X_cls_test)
```

```python
# Performance analysis
from sklearn.metrics import classification_report

report = classification_report(y_cls_test, y_cls_pred)
print(report)
```

```python
# Regression tree model
from sklearn.tree import DecisionTreeRegressor as DTR

dtr = DTR(min_samples_leaf=200)
dtr.fit(X_rgs_train, y_rgs_train)
```

```python
# Visualize the model
from sklearn import tree
import graphviz

dot_data = tree.export_graphviz(dtr, out_file=None, filled=True,
                                feature_names=predictors_list)
graphviz.Source(dot_data)
```
